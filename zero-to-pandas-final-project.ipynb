```python
analysis_df['Date']=pd.to_datetime(analysis_df.Date)
```

```python
analysis_df['year'] = pd.DatetimeIndex(analysis_df.Date).year
analysis_df['month'] = pd.DatetimeIndex(analysis_df.Date).month
analysis_df['weekday'] = pd.DatetimeIndex(analysis_df.Date).weekday
```

```python
analysis_df
```

```python
yearly_analysis_df=analysis_df[['Open',
                              'Close',
                              'High',
                              'Low',
                              'Previous',
                              'Change',
                              '%Change']].groupby(analysis_df.year).mean()
yearly_analysis_df
```

```python
monthly_analysis_df=analysis_df[['Open',
                                 'Close',
                                 'High',
                                 'Low',
                                 'Previous',
                                 'Change',
                                 '%Change']].groupby(analysis_df.month).mean()

monthly_analysis_df
```

```python
weekday_analysis_df=analysis_df[['Open',
                                 'Close',
                                 'High',
                                 'Low',
                                 'Previous',
                                 'Change',
                                 '%Change']].groupby(analysis_df.weekday).mean()
weekday_analysis_df
```

```python
analysis_df.drop('year',axis='columns',inplace=True)
analysis_df.drop('month',axis='columns',inplace=True)
analysis_df.drop('weekday',axis='columns',inplace=True)
```

I have created the new data set using the dates from the **analysis_df** dataframe.

**Next check values of other columns are right or wrong**:-

```python
analysis_df[analysis_df.Open==0.00]
```

The stocks of nifty can't be zero .so we have to change it to approximate value based on previous value and close value

```python
analysis_df.Open[analysis_df.Open==0.00]=analysis_df.Previous[analysis_df.Open==0.00]
```

```python
analysis_df.loc[2963]
```

Now You can see that the value that contain 0.00 has been changed based on some performance 

Since the values of nifty must not be 0 .So we also have to change High and Low values from 0 to approx value (we can't able to find the perfect value).

```python
(analysis_df.High-analysis_df.Close).mean()
```

```python
analysis_df.High[analysis_df.High==0.00]=analysis_df.Close[analysis_df.High==0.00]+0.781
```

```python
analysis_df.loc[2963]
```

```python
(analysis_df.Close-analysis_df.Low).mean()
```

```python
analysis_df.Low[analysis_df.Low==0.00]=analysis_df.Close[analysis_df.Low==0.00]-1.515
```

yeah! we have changed the totally wrong information in to approx info.

```python
analysis_df[round((analysis_df.Close-analysis_df.Previous),2)!=analysis_df.Change]
```

the value of 'Change' is also wrong in some of the places .So we have to have to correct it using < close value - previous value > 

```python
analysis_df["Change"]=round(analysis_df.Close-analysis_df.Open,2)
```

```python
analysis_df[(round((analysis_df.Change/analysis_df.Previous),4)!=analysis_df["%Change"])]
```

Wff !!!!!.... percentage change also varies we need to correct it . No problem we can do it in one line .

```python
analysis_df["%Change"]=round((analysis_df.Change/analysis_df.Previous),4)
```

### Q5. For which dates the data was not available?

```python
from datetime import datetime, timedelta
sdate =analysis_df.at[0,'Date']    # start date
edate = analysis_df.at[3060,'Date']   # end date

all_dates_df=pd.date_range(sdate,edate-timedelta(days=1))
```

```python
j=0
non_incuded_dates=pd.DataFrame(columns=['Date'])
for i in range( len(all_dates_df)):
    if all_dates_df[i]==analysis_df.Date[j]:
        j=j+1
    else:
        non_incuded_dates=non_incuded_dates.append({'Date':all_dates_df[i]},ignore_index=True)
```
